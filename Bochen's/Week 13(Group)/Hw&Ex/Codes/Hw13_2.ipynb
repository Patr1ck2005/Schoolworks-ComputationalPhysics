```python
import numpy as np
import matplotlib.pyplot as plt
import random

plt.rc('font', family='Times New Roman')


# 定义被积分函数
def fun(x):
    return 2 / (1 + x)


n = np.linspace(10, 1000000, 1000)

Value = []
Error = []
for N in n:
    N = int(N)
    X = []
    Y = []
    O = []
    # 生成自变量矩阵
    for j in range(N):
        X.append(random.random())
        O.append(0)
    # 生成因变量矩阵
    # 生成自变量矩阵
    for j in range(N):
        Y.append(fun(X[j]))

    # 计算平均值
    Mean = np.average(Y)
    v = Mean
    Value.append(v)
    Error.append(np.abs(v - 2 * np.log(2)))
    print('N:', N, "value:", v, "error:", np.abs(v - 2 * np.log(2)))
```

```python
log_n = np.log10(n)
error = np.log10(Error)
plt.figure(figsize=(6, 6))
plt.scatter(log_n, error, s=5, c='b', label='Error')
y = -1 / 2 * log_n - 2 / 3
plt.plot(log_n, y, c='r', label='Slope approaching curve')
plt.legend()
plt.title('Error')
plt.savefig(r'D:\FileArchiving\大二下\计算物理\瞎写的python\Week13\文档\误差.eps')

plt.figure(figsize=(6, 6))
plt.plot(n, Value, linewidth=1.5, c='b', label='Estimated value')
plt.title('Vaule')
plt.axhline(2 * np.log(2), linewidth=1.5, c='r', label='Theoretical value')
plt.legend()
plt.savefig(r'D:\FileArchiving\大二下\计算物理\瞎写的python\Week13\文档\函数.eps')
plt.show()
```
