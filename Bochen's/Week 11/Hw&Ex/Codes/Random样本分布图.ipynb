```python
import matplotlib.pyplot as plt
import numpy as np

# 生成1000个随机数作为x和y坐标
x = np.random.random(1000)
y = np.random.random(1000)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6, s=20)  # alpha控制透明度，s控制点的大小
plt.title('2D Distribution of 1000 Random Numbers')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.savefig(r'D:\FileArchiving\大二下\计算物理\瞎写的python\Week11\文档\Random样本分布图.eps')
plt.show()
```
