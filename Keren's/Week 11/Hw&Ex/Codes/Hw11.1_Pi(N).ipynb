```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数

def buffon(n, l, a):
    k = 0
    theta = np.random.uniform(0, np.pi, n)
    x = np.random.uniform(0, a / 2, n)
    for i in range(n):
        if x[i] <= l * np.sin(theta[i]) / 2:
            k += 1
    if k == 0:
        return 0
    pi = (2 * l * n) / (a * k)
    return pi

# 参数设定
a = 1
l = 0.5  # 选择一个中间值，这里假设l=0.5
num_simulations = 100  # 模拟次数
max_n = 100000  # 最大的投针次数

# 初始化数组来存储结果
n_values = np.arange(1000, max_n+1, 1000)
pi_estimates = np.zeros(len(n_values))

# 进行多次模拟
for j in range(num_simulations):
    for i, n in enumerate(n_values):
        pi_estimates[i] += buffon(n, l, a)

# 求平均值
pi_estimates /= num_simulations

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(n_values, pi_estimates, marker='o', linestyle='-', color='b', label='Estimated Pi')
plt.axhline(y=np.pi, color='r', linestyle='--', label='Exact Pi')
plt.xscale('log')
plt.title('Estimation of Pi vs Number of Needle Droppings')
plt.xlabel('Number of Needle Droppings (log scale)')
plt.ylabel('Estimated Value of Pi')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
